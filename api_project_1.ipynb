```python
#import requests
import time

# The public API endpoint for the model
API_URL = "https://huggingface.co"

# 🔒 SAFE FOR SHARING: Replace the placeholder below with your private token ONLY when running locally.
# Never commit your real token string to public repositories or share it in screenshots.
headers = {"Authorization": "Bearer YOUR_HF_API_KEY_PLACEHOLDER"}

def get_ai_response(prompt):
    payload = {
        "inputs": prompt,
        "parameters": {"max_new_tokens": 250, "do_sample": True, "temperature": 0.7}
    }

    while True:
        try:
            response = requests.post(API_URL, headers=headers, json=payload)

            # Catching authorization issues safely
            if response.status_code == 401:
                return "API Error: Please verify your Hugging Face API Token configuration."

            result = response.json()

            # Handling model cold-start/wakeup times gracefully
            if isinstance(result, dict) and "estimated_time" in result:
                print(f"⏳ Model is initializing... waiting {int(result['estimated_time'])} seconds...")
                time.sleep(5)
                continue

            if isinstance(result, list) and len(result) > 0:
                return result[0]['generated_text']
            elif isinstance(result, dict) and 'error' in result:
                return f"API Error: {result['error']}"
            return "Error: Unexpected response format from API."

        except Exception as e:
            return f"Connection Error: {str(e)}"

# --- Main Program Execution ---
topic = input("Enter your study topic: ")

print("\n🚀 Contacting the AI engine... please hold on...")

summary_prompt = f"Explain in simple notes: {topic}"
quiz_prompt = f"Generate 3 MCQs with answers on: {topic}"
key_prompt = f"Give key points for: {topic}"

summary = get_ai_response(summary_prompt)
quiz = get_ai_response(quiz_prompt)
key_points = get_ai_response(key_prompt)

print("\n📘 SUMMARY:\n", summary)
print("\n🧠 QUIZ:\n", quiz)
print("\n📌 KEY POINTS:\n", key_points)
```
